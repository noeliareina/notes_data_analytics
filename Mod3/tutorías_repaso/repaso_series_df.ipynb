# DATAFRAMES

```python
import pandas as pd

data = {
    "Nombre": ["Ana", "Luis", "Marta", "Carlos", "Sofía", "Pedro"],
    "Edad": [25, 30, 45, 28, None, 33],
    "Ciudad": ["Madrid", "Barcelona", "Sevilla", "Madrid", "Barcelona", "Sevilla"],
    "Salario": [3000, 2500, 4000, None, 2700, 3200]
}

df = pd.DataFrame(data)
```

```python
# 1.¿Cuántos valores nulos hay en Edad y Salario?
df[["Edad", "Salario"]].isnull().sum()
```

```python
# 2. ¿Cuántas personas hay por ciudad?
df["Ciudad"].value_counts()
```

```python
#3. ¿Cuál es la media de los salarios?
df["Salario"].mean()
```
