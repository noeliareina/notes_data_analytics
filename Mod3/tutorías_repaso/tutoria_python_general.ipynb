````python
def filtrar_mayores(lista_usuarios, edad_minima):
    resultado = []
    for usuario in lista_usuarios:
        if usuario["edad"] >= edad_minima:
            resultado.append(usuario["nombre"])
    return resultado
````

````python
# CONTEO
notas = [4.5, 7.8, 9.1, 3.2, 5.0, 8.6, 10]
def clasificar_notas (lista_notas):
    contador = {"suspensos": 0, "aprobados": 0, "sobresalientes": 0}

    for nota in lista_notas:
        if nota < 5:
            contador["suspensos"] += 1
        elif 5 <= nota < 9:
            contador["aprobados"] += 1
        else:
            contador["sobresalientes"] += 1
    return contador
notas_rdos = clasificar_notas(notas)
print(f"los resultados del examen son {notas_rdos}" )
````
