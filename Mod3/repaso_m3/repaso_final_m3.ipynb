```python
# Creamos una función que nos permita aplicar una iteración de remplazo de sintaxis

# x = np.nan
# Try Except por si nos encontramos con eso porque a eso no se puede hacer .replace
```

```python
def cambiar_float(dato):
	try:
		return float(dato.replace(",", "."))
	except:
		return np.nan

x = "2,3"

cambiar_float(x)
```

```python
columnas = ["price", "reference_price"] #cambio de title del 0 por el 1 índice

for col in columnas:
    df_unido[col] = df_unido[col].apply(cambiar_float) #aplicamos la función a cada columna con el bucle de iteración
```

```python
def extraer_categorias(url):
    partes = url.split("/")  # Dividir la URL por "/"  
    try:
        categoria = partes[4]  # Ajusta este índice según la estructura de tu URL, el elemento 4
        subcategoria = partes[5] if len(partes) > 5 else None  # Ajusta este índice, el elemento 5 si la longitud es > 5, si tuviera subcategoría
        return pd.Series([categoria, subcategoria])
    except IndexError:
        return pd.Series([None, None]) # La función nos devuelve una serie con dos elementos, cat y subcat
    
# Lambdas: cuando las funciones reciben 2 parámetros
# Lambdas también cuando devuelve 2 elementos la función
```

```python
df_unido["url"].apply(extraer_categorias)
```

```python
## otra función para obtener categorias y subcategorias
def sacar_info(x):
    if "al-dia" in x:
        return "frescos", x.split("/")[5]

    elif "bulto-mixto" in x or "peso" in x or "caducidad" in x or "clasificacion-de-categorias" in x:
        return x.split("/")[5], x.split("/")[6]
        
    else:
        return x.split("/")[4], x.split("/")[5]
```

```python
sacar_info('https://www.dia.es/compra-online/cuidado-personal/bano-e-higiene-personal/cuidado-del-cabello/p/283797')
```

```python
df_unido[["categoria", "subcategoria"]] = df_unido.apply(lambda x: extraer_categorias(x["url"]), axis=1, result_type="expand")
```
