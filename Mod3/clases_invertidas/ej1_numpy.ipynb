# 5
Crea un array bidimensional con 20 elementos aleatorios entre 0 y 50 con el m√©todo que prefieras y:

- Crea una mascara o filtro donde los valores mayores a 30 y menores a 10 se reemplace por "objetivo" y el resto por "no es mi objetivo".

- Los valores de mi array bidimensional multiplicalos por el escalar 10.

```python
array = np.random.randint(0, 51, (4, 5))
print("Array original:")
print(array)

# Crear una m√°scara para los valores mayores a 30 y menores a 10
mascara = (array > 30) | (array < 10)

# Crear un array de etiquetas basado en la m√°scara
etiquetas = np.where(mascara, "objetivo", "no es mi objetivo")
print("\nM√°scara aplicada:")
print(etiquetas)

# Multiplicar el array por 10
array_escalado = array * 10
print("\nArray multiplicado por 10:")
print(array_escalado)
```

# 6
Crea un nuevo array de dos dimensiones con la misma forma que el anterior y con el m√©todo que prefieras y realiza lo siguiente:

- Suma el primer array con el segundo.

- Aplicar la potencia para el primer array elevado al segundo.

```python
import numpy as np

# Crear un array bidimensional de 4x5 con valores aleatorios entre 0 y 50
array = np.random.randint(0, 51, (4, 5))
print("Array original:")
print(array)

# Crear una m√°scara para los valores mayores a 30 y menores a 10
mascara = (array > 30) | (array < 10)

# Crear un array de etiquetas basado en la m√°scara
etiquetas = np.where(mascara, "objetivo", "no es mi objetivo")
print("\nM√°scara aplicada:")
print(etiquetas)

# Multiplicar el array por 10
array_escalado = array * 10
print("\nArray multiplicado por 10:")
print(array_escalado)

# Crear un segundo array con la misma forma que el primero
array2 = np.random.randint(0, 51, (4, 5))
print("\nSegundo array:")
print(array2)

# Sumar el primer array con el segundo
suma_arrays = array + array2
print("\nSuma de los dos arrays:")
print(suma_arrays)

# Aplicar la potencia del primer array elevado al segundo
potencia_arrays = np.power(array, array2)
print("\nPrimer array elevado al segundo:")
print(potencia_arrays)
```

# 7
Crea un array, puede ser bidimensional! Eso si, los valores tienen que estar entre 0-100

- Calcula:

        - La media de los elementos

        - La media de las columnas y redond√©alo a dos decimales

        - El valor m√°ximo de cada fila

        - El valor m√≠nimo de cada fila

        - Hay algun valor que sea mayor que 90 en tu array

        - Donde los podemos encontrar, cu√°les son sus √≠ndices

        - Ordena el array de mayor a menor

üìù NOTA En este ejercicio os hemos pedido algunos de los m√©todos, pero sentiros libres de practicar otros para ir asentando los conocimientos.

```python
# Crear un array bidimensional de 4x5 con valores aleatorios entre 0 y 100
array = np.random.randint(0, 101, (4, 5))
print("Array original:")
print(array)

# Calcular la media de los elementos
total_media = np.mean(array)
print("\nMedia de los elementos:", total_media)

# Calcular la media de las columnas y redondear a dos decimales
media_columnas = np.round(np.mean(array, axis=0), 2)
print("\nMedia de las columnas:", media_columnas)

# Obtener el valor m√°ximo de cada fila
max_filas = np.max(array, axis=1)
print("\nValor m√°ximo de cada fila:", max_filas)

# Obtener el valor m√≠nimo de cada fila
min_filas = np.min(array, axis=1)
print("\nValor m√≠nimo de cada fila:", min_filas)

# Comprobar si hay alg√∫n valor mayor que 90
hay_mayor_90 = np.any(array > 90)
print("\n¬øHay alg√∫n valor mayor que 90?:", hay_mayor_90)

# Obtener los √≠ndices de los valores mayores a 90
indices_mayor_90 = np.argwhere(array > 90)
print("\n√çndices de los valores mayores a 90:", indices_mayor_90)

# Ordenar el array de mayor a menor
array_ordenado = np.sort(array, axis=None)[::-1]
print("\nArray ordenado de mayor a menor:")
print(array_ordenado)
```
