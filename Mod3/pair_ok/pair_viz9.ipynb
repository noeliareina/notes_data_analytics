4. Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil.

```python
import pandas as pd

# Función para clasificar según los valores de mortalidad infantil
def clasificar_mortalidad(valor):
    if valor <= 10:
        return "Baja"
    elif valor <= 30:
        return "Moderada"
    else:
        return "Alta"

# Aplicamos la función a la columna 'infant_mortality'
df['infant_mortality_category'] = df['infant_mortality'].apply(clasificar_mortalidad)
```
