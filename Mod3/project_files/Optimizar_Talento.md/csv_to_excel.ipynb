```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Cargar el archivo CSV en un DataFrame
df = pd.read_csv("ejemplo_excel_graficos.csv")

# 2️⃣ Ver las primeras filas para entender los datos
print(df.head())

# 3️⃣ Crear un gráfico de barras
plt.figure(figsize=(8,5))
sns.barplot(x="total_peliculas_alquiladas", y="nombre_categoría", data=df)
plt.title("Gráfico de Barras")
plt.xticks(rotation=45)
plt.show()
```
