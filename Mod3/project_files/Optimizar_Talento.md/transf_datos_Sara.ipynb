## Creo función para valores nulos y vacíos

```python
def filtrado_nulos (df,columna):
    
    df_nulos =df[df[columna].isnull()]
    df[columna].isnull().sum()
    
    if df_nulos.empty: 
        print (f'No hay valores nulos en la columna {columna}')

    else:
        print(f'El número total de nulos de la columna es de {df[columna].isnull().sum()}')
        print(f'Los valores nulos de la columna {columna} son:')
        print(df_nulos[columna])
```

```python
def filtrado_vacios (df,columna):
    
    df_vacios = df[df[columna] == " "]
    
    if df_vacios.empty:
        print (f'No hay valores vacíos en la columna {columna}')

    else:
        print(f'Los valores vacíos de la columna {columna} son:')
        print(df_vacios[columna])
```

¿Qué significa dtype('O')?

    dtype es una abreviatura de Data Type (tipo de dato), que es una propiedad de los arrays y objetos en pandas que especifica qué tipo de datos están almacenados.
    'O' hace referencia a "Object", un tipo genérico que pandas utiliza para almacenar datos que no pertenecen a tipos numéricos específicos o tipos más estructurados (como enteros, flotantes, fechas, etc.).

En el contexto de un DataFrame de pandas, si una columna tiene dtype('O'), significa que los valores de esa columna son objetos y no un tipo de dato numérico o una fecha, normalmente implicando que son cadenas de texto (strings) o valores mixtos.
