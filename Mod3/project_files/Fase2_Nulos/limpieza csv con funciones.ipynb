```python
# Paso 1: Convierto a string y quito símbolos con .replace normal
df['salary'] = df['salary'].astype(str) #Convierte todos los valores de la columna a texto (str), necesario para usar .str.replace() aunque haya numeros o NaN.
df['salary'] = df['salary'].str.replace('$', '') # Quita el símbolo dólar ($) de todos los valores.
df['salary'] = df['salary'].str.replace(',', '.') #Reemplaza la coma decimal por un punto decimal.

# Paso 2: Convierto a float (NaN donde no se pueda)
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

# Paso 3: Reemplazo NaN por None (para SQL)
df['salary'] = df['salary'].where(pd.notnull(df['salary']), None)
```

```python
# FUNCION PARA LIMPIAR COLUMNAS OBJECT Y CONVERTIR A NUMERICAS:

def clean_numeric_column(df, column):  # df:dataframe, column: el nombre de la columna a limpiar.
    # Paso 1: convierto a texto y limpio símbolo
    df[column] = df[column].astype(str) # Convierto toda la columna a str, porque .str.replace() solo funciona con texto.
    df[column] = df[column].str.replace('$', '')
    df[column] = df[column].str.replace(',', '.')

    # Paso 2: convertir a float
    # errors = coerce -> Estoy forzando ->“Intenta convertir todo a número. Si encuentras algo que no se puede convertir (como texto raro o vacío), en vez de dar error, pon un NaN”.
    df[column] = pd.to_numeric(df[column], errors='coerce') # pd.to_numerics es una funcion de Pandas que  convierte strings en numeros int/float

    # Paso 3: convertir NaN a None para SQL
    df[column] = df[column].where(pd.notnull(df[column]), None) # si el valor es NaN, cambialo a None, sino dejalo tal cual.

    return df  # devuelvo el df modificado.
```

```python
# creo lista con las columnas a las que le voy a aplicar la funcion que he creado.

columnas_a_limpiar = ['monthlyincome', 'sameasmonthlyincome', 'monthlyrate','totalworkingyears', 'performancerating', 'worklifebalance','salary']
```

```python
# hago un bucle for para aplicar la funcion a la lista de columnas que he definido.

for col in columnas_a_limpiar:
    clean_numeric_column(df, col)
```

```python
# ya he importado la libreria, hago una funcion para limpiar la columna age:
# necesito pasarlo todo a texto/string primero.

def convertir_edad(valor):  # Definimos la función. Le pasaremos una edad, que puede ser: un numero ("42"), un string o un Nulo (NaN)
    if pd.isnull(valor):    # Si el valor es nulo (NaN), lo dejamos como None. 
        return None
    valor = str(valor).strip().lower() # str(valor) lo convierte a texto/string por si viene como numero, con .strip() elimino posibles espacios y .lower()lo paso todo a minusculas.
    try:
        return int(valor) if valor.isdigit() else w2n.word_to_num(valor)  # si el valor es .isdigit() significa que es un numero tipo "42" y uso INT para convertirlo, si no es numero sino letra (forty-two) uso w2n para traducirlo.
    except:
        return None  # Por si hay algún texto no válido


#con el try me aseguro de cambiar a INT si es un "35" y me de 35 y si esta escrito "twenty" uso w2n para que me de 20 en INT
```

```python
df["age"] = df["age"].apply(convertir_edad)

# con apply aplico la funcion que acabo de hacer a cada fila de la columna AGE para que realice los cambios a INT. 
# El resultado se guarda en la misma columna.
```

RESUMEN DE CAMBIOS DE AGE:
- .strip()
Elimina espacios en blanco al principio y al final de un string.
Ejemplo: ' hola '.strip() -> 'hola'
- .isdigit()
Devuelve True si todos los caracteres del string son dígitos (números enteros positivos).
Ejemplo: '42'.isdigit() -> True, 'forty-two'.isdigit() -> False.  No detecta decimales, negativos ni espacios. 
- .apply()
Aplica una función a cada valor de una columna (o fila si se usa con axis=1).
El resultado se guarda en la columna.

```python
# Paso 1: Convierto a texto, quito espacios y paso a minúsculas
df['remotework'] = df['remotework'].astype(str).str.strip().str.lower()
```

```python
# Paso 2: Hago funcion para unificar todos los valores similares a "yes" y "no"
def estandarizar_remotework(valor):
    if valor in ['yes', '1', 'true']:
        return 'yes'
    elif valor in ['no', '0', 'false']:
        return 'no'
    else:
        return None  # Por si hay valores raros o vacíos

df['remotework'] = df['remotework'].apply(estandarizar_remotework)
```

```python
columnas_binarias = ['remotework', 'overtime', 'attrition', 'over18']
```

```python
def clean_binary_column(df, column):
    # Paso 1: convertir a texto, quitar espacios y pasar a minúsculas
    df[column] = df[column].astype(str).str.strip().str.lower()

    # Paso 2: estandarizar valores (sí/no)
    def estandarizar(valor):
        if valor in ['yes', '1', 'true']:
            return 'yes'
        elif valor in ['no', '0', 'false']:
            return 'no'
        else:
            return None  # Valor inesperado

    # Aplicamos la estandarización
    df[column] = df[column].apply(estandarizar)
    
    return df
```
