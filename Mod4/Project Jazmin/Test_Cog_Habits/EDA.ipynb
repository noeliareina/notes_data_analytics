6. Identificamos los valores únicos en todas las columnas para detectar inconsistencias o errores tipográficos.

```python
#Lo hacemos en bucle para revisar todas las columnas
for columna in df:
    print(f'La columna {columna} tiene estos valores únicos:')
    print(df[columna].unique())
    print('----------------------------') 
```

7. Verificamos si hay filas duplicadas.

```python
duplicados = df.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados}")
# Aquí con .sum() sabemos la cantidad, pero queremos ver el detalle. Para nosotras solo habría duplicados en el 'employeenumber' o columnas.
```

```python
df[df.duplicated(keep=False)].sort_values('employeenumber')
# Aquí vemos detalladamente las filas que tienen duplicidades en 'employeenumber': 128/2 = 64, todo lo demás es igual
```

### Conclusiones del análisis exploratorio:

- Hay 64 filas duplicadas, ¿las debemos eliminar ya, esperamos al final? Hemos comprobado que esos duplicados son de employeenumber y todo los demás datos son iguales, por lo que decidimos eliminarlas.
- Hay valores nulos, ¿los convertimos todos a None o esperamos?
