```python
import pandas as pd
import time
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

df = pd.read_csv('Grabacion.csv')
df.head(3)
```

```python
df.drop(columns=['0'], inplace=True)
df.head(3)
```

Tener en cuenta que no puede haver t√≠tulos repetidos

```python
# üóÇÔ∏è Ruta donde se descargan los archivos
downloads_folder = os.path.expanduser("~/Downloads")

def limpiar_y_renombrar(titulo_deseado):

    # Buscar archivos reci√©n descargados
    archivos = os.listdir(downloads_folder)

    mp4_file = None
    for archivo in archivos:
        if archivo.endswith("1080.mp4") or archivo.endswith("1440.mp4"):
            mp4_file = os.path.join(downloads_folder, archivo)
        elif archivo.endswith((".txt", ".m4a")):
            os.remove(os.path.join(downloads_folder, archivo))  # Eliminar los que no nos interesan

    if mp4_file:
        nuevo_nombre = f"{titulo_deseado}.mp4"
        destino = os.path.join(downloads_folder, nuevo_nombre)
        os.rename(mp4_file, destino)
        print(f"‚úÖ Renombrado a: {nuevo_nombre}")
    else:
        print("‚ö†Ô∏è No se encontr√≥ archivo .mp4")
```

```python
options = Options()
options.add_experimental_option("prefs", {
    "profile.default_content_setting_values.automatic_downloads": 1  # ‚úÖ Permitir m√∫ltiples descargas
})

archivo = 30

for link in df['Link'].tolist()[archivo:]:
    
    driver = webdriver.Chrome(options=options) 
    driver.get(link)

    time.sleep(4)

    accept_cookies = driver.find_element(By.CSS_SELECTOR, '#onetrust-accept-btn-handler')
    accept_cookies.click()

    time.sleep(2)

    download = driver.find_element(By.CSS_SELECTOR, '#app > header > div.header-right > div > a')
    download.click()

    time.sleep(30)

    titulo = df['T√≠tulo'].tolist()[archivo]
    titulo_clean = f'{archivo}_' + titulo.lower().replace(',', '').replace('/', ' ').replace('.', '').replace(':', '').replace('¬¥', '').replace(' ', '_')
    limpiar_y_renombrar(titulo_clean)
    

    try: 
        next_recording = driver.find_element(By.CSS_SELECTOR, '#vjs_video_3 > div.vjs-control-bar > div.vjs-multiple-clip-control > div.vjs-multiple-clip-control-button.button-next > button')
        next_recording.click()

        time.sleep(3)

        download = driver.find_element(By.CSS_SELECTOR, '#app > header > div.header-right > div > a')
        download.click()

        time.sleep(30)

        titulo_part2 = titulo_clean + 'Part2'
        limpiar_y_renombrar(titulo_part2)

        
    except:
        pass

    archivo += 1

    driver.close()

    
```
