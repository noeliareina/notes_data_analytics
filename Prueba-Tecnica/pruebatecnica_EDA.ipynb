Data preparation: 
- IN THE END: Column names start with uppercase, we put all column names in lowercase. 
- Convert tweet_id from object to int

```python
# 1)Rename the columns to remove spaces and special characters
# Rename the columns to lowercase
df.columns = df.columns.str.lower().str.replace(' ', '_') # remove spaces and special characters from the column names
```

```python
## Necessary to convert to string for consistency
df["tweet_id"] = df["tweet_id"].astype(str)
df["user_id"] = df["user_id"].astype(str)
```

```python
# Convert tweet_id to numeric, removing quotes and stripping whitespace
df["tweet_id"] = pd.to_numeric(
    df["tweet_id"].str.replace('"', "", regex=False).str.strip().str.lstrip("-"),
).astype("Int64")
```

```python
df["user_id"] = pd.to_numeric(
    df["user_id"].str.replace('"', "", regex=False).str.strip(),
    errors="coerce"
).astype("Int64")
```

```python
df["tweet_id"] = df["tweet_id"].abs()
df["user_id"] = df["user_id"].abs()
```

```python
def extraer_pais(location, mapping):
    if not isinstance(location, str):
        return 'other'

    location = location.lower()
    for key, iso in mapping.items():
        if key in location:
            return iso
    return 'other'
```

```python
df['pais_iso2'] = df['tweet_location_clean'].apply(lambda x: extraer_pais(x, country_mapping))
```
