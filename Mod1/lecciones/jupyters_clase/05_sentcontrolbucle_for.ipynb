```python
# 14. Pide al usuario un numero
#  entero positivo y suma los numeros 
# pares e impares por separado hasta ese numero:
numero = int(input("ingresa un número entero positivo: "))

for i in range(1, numero+1): #+1, porque no nos incluye el último

    if i % 2 == 0:
        print(i, "par")

    else:
        print(i, "impar")
```

```python
numero = int(input("ingresa un número entero positivo: "))
pares = []
impares = []
for i in range(1, numero+1):

    if i % 2 == 0:
        pares.append(i)

    else:
        impares.append(i)
print(sum(pares), pares)
print(sum(impares), impares)
```

```python
numero = int(input("ingresa un número entero positivo: "))
pares = 0
impares = 0
for i in range(1, numero+1):

    if i % 2 == 0:
        pares += i # IMPORTANTE

    else:
        impares += i

print(pares)
print(impares)
```

```python
for elemento in compra.values():
    print(elemento)
```

```python
unidades = 0
for elemento in compra.values():
    unidades += elemento

    print(unidades)
```
