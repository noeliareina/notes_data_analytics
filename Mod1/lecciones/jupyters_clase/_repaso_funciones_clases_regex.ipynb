```python
class Ecomerce:
    def __init__(self, nombre):
        
        self.nombre = nombre
        self.clientes = {}
        self.productos = {}
        self.ventas = []

    def agregar_producto(self, id_producto, nombre, precio, stock):
        if id_producto in self.productos:
            print("Ya existe")
        else:
            self.productos[id_producto] = [nombre, precio, stock]
            print("No existe, lo hemos insertado")

    def eliminar_producto(self, id_producto):
        if id_producto in self.productos:
            self.productos.pop(id_producto)
            print(f"el producto {id_producto}, ha sido eliminado")
        else:
            print("Ese producto no existe")    

    def agregar_cliente(self, id_cliente, nombre, email):

        if id_cliente in self.clientes:

            print(f"El cliente con id {id_cliente} y nombre {nombre} ya está registrado")
        
        else:

            self.clientes[id_cliente] = [nombre, email]
            print(f"No existe, el cliente {id_cliente} ha sido agregado")

tienda = Ecomerce("La tendeta")
tienda.agregar_producto(12, "Patatas", 25, 100)
tienda.agregar_producto(13, "cebollas", 30, 130)
tienda.eliminar_producto(12)

tienda.agregar_cliente("22c", "Paco Pill", "pp@pill.com")
```
