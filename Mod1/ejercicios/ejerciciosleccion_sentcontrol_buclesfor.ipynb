1. Usando la siguiente lista:
Crea tres listas nuevas, donde:

- En la primera tengamos temperaturas entre 0 - 10, llama a la nueva lista bajas_temp.

- En la segunda tengamos temperaturas entre 11- 20, llama a la nueva lista medias_temp.

- En la tercera tengamos temperaturas por encima de 20, llama a la nueva lista altas_temp

```python
temperaturas = [2,45,67,23, 45,66,12,1,34,67,99,98,67,64, 44, 100]
print("La lista de temperaturas es:", temperaturas)
print("-----------")
# ¿Cómo saco 3 listas de la variable "temperaturas" con una condición cada una?
# creamos listas vacías

```

```python
bajas_temp = []
medias_temp =[]
altas_temp = []
print("----------------")

for temp in temperaturas:
    #sacamos por pantalla las temperaturas que estamos iterando.
    if 0 <= temp <= 10: 
        bajas_temp.append(temp)
    elif 11 <= temp <= 20:
        medias_temp.append(temp)
    elif temp > 20: 
        altas_temp.append(temp)

# Imprimimos las listas clasificadas
print("Temperaturas bajas (0-10):", bajas_temp)
print("Temperaturas medias (11-20):", medias_temp)
print("Temperaturas altas (>20):", altas_temp)

#ordenar
print(sorted(bajas_temp))
print(sorted(medias_temp))
print(sorted(altas_temp))

  
```

2. Las temperaturas están grados centígrados. Para cada una de las listas creadas en el ejercicio anterior, convertid las temperaturas a grados fahrenheit.

```python
# la relacion de grados celsius y fahrenheit es:
#(0 °C × 9 / 5) + 32 = 32 °F
bajas_temp_fahrenheit = []
medias_temp_fahrenheit = []
altas_temp_fahrenheit = []

# Función para convertir Celsius a Fahrenheit
def celsius_a_fahrenheit(lista):
    return [(temp * 9/5) + 32 for temp in lista]  # Lista por comprensión
# Tras hacer la transformación teniendo en cuenta cada temperatura concreta de la lista temperaturas, añadimos contenido a nuevas listas
bajas_temp_fahrenheit.append(celsius_a_fahrenheit(bajas_temp))
print("La lista de temp bajas en fahrenheit es", bajas_temp_fahrenheit)

medias_temp_fahrenheit.append(celsius_a_fahrenheit(medias_temp))
print("La lista de temp medias fahrenit es", medias_temp_fahrenheit)

altas_temp_fahrenheit.append(celsius_a_fahrenheit(altas_temp))
print("La lista de temp bajas en fahrenheit es", altas_temp_fahrenheit)
```
