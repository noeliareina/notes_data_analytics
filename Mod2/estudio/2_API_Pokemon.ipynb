# *FUNCION PARA LLAMAR A LA API*

```python
def llamar_api(endpoint):
    llamada = requests.get(endpoint)

    print(f"La llamada a la API nos ha dado una respuesta de tipo {llamada.status_code}")
    if llamada.status_code != 200:
        print(f"El motivo por el que la llamada fall√≥ es {llamada.reason}")
    else:
        return llamada.json()
```

```python
url_black = "https://pokeapi.co/api/v2/pokemon-color/black/"
response_black = llamar_api(url_black)
response_black
```

```python
url_purple = "https://pokeapi.co/api/v2/pokemon-color/purple/"
response_purple = llamar_api(url_purple)
```

```python
diccionario_resultados_color = []
```

```python
def limpiar_color_pokemon(resultados_json, diccionario_resultados_color):
    
    diccionario_resultados_color["color"].append(resultados_json["name"])
    
    lista_especies = []
    
    for especie in resultados_json["pokemon_species"]:
        lista_especies.append(especie["name"])
    
    diccionario_resultados_color["especies"].append(lista_especies)

    return diccionario_resultados_color
```

```python
dict_color = {"color": [], "especies": []}
```

```python
diccionario_color = limpiar_color_pokemon(response_black, dict_color)
diccionario_color
```

```python
diccionario_color_morado = limpiar_color_pokemon(response_purple, dict_color)
diccionario_color_morado
```

```python
lista_colores = ["black", "red", "green", "purple", "brown", "yellow", "blue", "orange"]

diccionario_todos_color = {"color": [], "especies": []}

for color in lista_colores:
    try:
        url_color = f"https://pokeapi.co/api/v2/pokemon-color/{color}/"
        
        json_color= llamar_api(url_color)

        diccionario_todos_color = limpiar_color_pokemon(json_color, diccionario_todos_color)
        # LA RESPUESTA es json_color u el dict de todos

    except:
        print(f'estamos teniendo problemas con color {color}')
```

```python
def clean_data_pokemon (datos_json, dict_pokemon):

    dict_pokemon["name"].append(datos_json["name"])

    dict_pokemon["height"].append(datos_json["height"])

    dict_pokemon["weight"].append(datos_json["weight"])

    dict_pokemon["types"].append(datos_json['types'][0]['type']['name'])

   #dict_pokemon["abilities"].append(abilities_pikachu_list) - ASI NO

    abilities_pokemon_list = []

    for ability in datos_json["abilities"]:

        abilities_pokemon_list.append(ability["ability"]["name"])

    dict_pokemon["abilities"].append(abilities_pokemon_list)
    
    location_pokemon = datos_json["location_area_encounters"]
    response_loc_pokemon = llamar_api(location_pokemon)
    
    pokemon_loc_list= []
    for ubicacion in response_loc_pokemon:
        pokemon_loc_list.append(ubicacion["location_area"]["name"])
    print(response_loc_pokemon)
    dict_pokemon["locations"] .append(pokemon_loc_list)

    return dict_pokemon
```

```python
dict_pikachu = {"name": [],
                "height": [],
                "weight": [],
                "types": [],
                "abilities": [],
                "locations":[]}
```

```python
resultado_dict_pikachu = clean_data_pokemon(response_pikachu, dict_pikachu)
```

## QUEREMOS LA FUNCION PARA +1 POKEMON

```python
lista_pokemons = ['Pikachu','blitzle','mawile', 'spoink']

dict_all_pokemons = {"name": [],
                "height": [],
                "weight": [],
                "types": [],
                "abilities": [],
                "locations":[]}
# creamos FOR LOOP para cada pokemon y lista
for pokemon in lista_pokemons:
    url_pokemon = f"https://pokeapi.co/api/v2/pokemon/{pokemon.lower()}"
    print(url_pokemon)

    results_api = llamar_api(url_pokemon)
    dict_all_pokemons = clean_data_pokemon (results_api, dict_all_pokemons)
```
