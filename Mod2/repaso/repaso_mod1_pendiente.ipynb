# Listas

12. Acceder a la segunda lista de frutas: Crea una variable llamada segunda_lista y asigna la segunda lista de frutas a esta variable.

14. Acceder a la última fruta de la segunda lista: Crea una variable llamada ultima_fruta y asigna la última fruta de la segunda lista de frutas a esta variable.

17. Acceder a la longitud de la segunda lista de frutas: Crea una variable llamada longitud_segunda_lista y asigna la longitud (número de elementos) de la segunda lista de frutas a esta variable.

18. Acceder a una porción de la segunda lista (slicing): Crea una variable llamada porcion_segunda_lista y asigna los dos primeros elementos de la segunda lista de frutas a esta variable utilizando el slicing.

19. Cambiar la segunda fruta de la segunda lista: Cambia la segunda fruta de la segunda lista de frutas a "ciruela".

20. Agregar una nueva fruta a la primera lista: Agrega una nueva fruta, "kiwi", a la primera lista de frutas.


# Tuplas
Crear una tupla: Crea una tupla llamada mi_tupla con algunos elementos de tu elección.

Acceder al primer elemento: Accede al primer elemento de mi_tupla y guárdalo en una variable llamada primer_elemento.

Acceder al último elemento: Accede al último elemento de mi_tupla y guárdalo en una variable llamada ultimo_elemento.

Acceder a un rango de elementos: Utilizando el slicing, crea una nueva tupla llamada sub_tupla que contenga los elementos desde el segundo hasta el cuarto de mi_tupla.

Concatenar dos tuplas: Crea otra tupla llamada otra_tupla y concaténala con mi_tupla para formar una nueva tupla llamada tupla_concatenada.

Encontrar la longitud: Encuentra la longitud de tupla_concatenada y guárdala en una variable llamada longitud.

Buscar un elemento: Verifica si un elemento específico está presente en mi_tupla, por ejemplo, "manzana".

Contar elementos: Cuenta cuántas veces aparece un elemento específico en mi_tupla, por ejemplo, cuántas veces aparece "pera".

Desempaquetar una tupla: Si tienes una tupla de nombres (nombre, apellido), desempaqueta los valores en dos variables separadas.

Convertir una lista en tupla: Crea una lista de números y conviértela en una tupla llamada tupla_desde_lista.

# Diccionarios
Crear un diccionario: Crea un diccionario vacío llamado mi_diccionario con una key llamada apellido y que su value sea López.

Agregar elementos: Agrega una clave-valor al diccionario mi_diccionario. Por ejemplo, agrega la clave "nombre" con el valor "Juan".

Acceder a un valor: Accede al valor correspondiente a la clave "nombre" en mi_diccionario.

Modificar un valor: Cambia el valor de la clave "nombre" a "Ana".

Eliminar un elemento: Elimina la clave "nombre" de mi_diccionario.

Verificar existencia: Verifica si la clave "apellido" existe en mi_diccionario.

Obtener todas las claves: Obtén una lista de todas las claves en mi_diccionario.

Obtener todos los valores: Obtén una lista de todos los valores en mi_diccionario.

Recorrer el diccionario: Usa un bucle for para recorrer y mostrar todas las claves y valores en mi_diccionario.

Longitud del diccionario: Calcula la longitud (cantidad de elementos) de mi_diccionario.

Copiar un diccionario: Crea una copia de mi_diccionario llamada copia_diccionario.

Eliminar elementos: Elimina todos los elementos de copia_diccionario.

Combinar diccionarios: Crea otro diccionario llamado diccionario_combinado y combina mi_diccionario y copia_diccionario en él.

Obtener valor predeterminado: Intenta acceder a la clave "edad" en mi_diccionario. Si no existe, muestra el valor predeterminado "Desconocida".

Clave máxima: Encuentra la clave con el valor máximo en un diccionario que contenga pares clave-valor numéricos.


# Sets
Crear un conjunto vacío: Crea un conjunto llamado mi_set que esté vacío.

Agregar elementos a un conjunto: Agrega los números 1, 2 y 3 al conjunto mi_set.

Eliminar elementos de un conjunto: Elimina el número 2 del conjunto mi_set.

Verificar pertenencia: Comprueba si el número 3 está en mi_set y almacena el resultado en una variable llamada existe.

Operaciones de conjuntos: Crea dos conjuntos set1 y set2 con números y encuentra la intersección de ambos conjuntos.

Unión de conjuntos: Combina set1 y set2 para formar un nuevo conjunto que contenga todos los elementos de ambos.

Diferencia de conjuntos: Encuentra los elementos que están en set1 pero no en set2.

Comparación de conjuntos: Comprueba si set1 es un subconjunto de set2 y almacena el resultado en una variable llamada es_subconjunto.

Eliminar elementos duplicados: Dada una lista mi_lista con elementos duplicados, conviértela en un conjunto para eliminar los duplicados.

Comprobar igualdad de conjuntos: Compara si set1 y set2 son iguales y almacena el resultado en una variable llamada son_iguales.


# Sentencias de control
Clasificación de triángulos: Pide al usuario que ingrese las longitudes de tres lados de un triángulo. Determina si el triángulo es equilátero (todos los lados iguales), isósceles (dos lados iguales), o escaleno (ningún lado igual).

Conversión de temperatura: Crea un programa que permita al usuario convertir temperaturas entre Celsius y Fahrenheit. Pide al usuario que ingrese la temperatura y la unidad (C o F) y realiza la conversión.

Cajero automático: Simula un cajero automático que permite a los usuarios verificar su saldo, depositar y retirar dinero. El programa debe mantener un registro del saldo actual.

Calculadora de descuento: Pide al usuario que ingrese el precio original de un producto y el porcentaje de descuento. Calcula el precio final después del descuento y muestra el ahorro.

Cálculo de promedio de calificaciones: Pide al usuario que ingrese una serie de calificaciones y calcule el promedio. También muestra si el estudiante aprobó (promedio mayor o igual a 5) o no. Pistas:

Solicita al usuario que ingrese 4 notas de calificaciones que quiere promediar.

Convierte el string en una lista de números.

Calcula el promedio de las calificaciones ingresadas.

Muestra el resultado del promedio con dos decimales.

Determina si el estudiante aprobó o no. Un estudiante se considera aprobado si su promedio es mayor o igual a 5.

Generador de contraseñas seguras: Crea un programa que genere una contraseña segura. El usuario debe especificar la longitud deseada y si desea incluir números, letras mayúsculas, minúsculas y caracteres especiales. Pistas:

Pregunta al usuario cuántos caracteres debe tener la contraseña que desea generar.


Luego, consulta al usuario si quiere incluir números, letras mayúsculas, letras minúsculas y caracteres especiales en su contraseña.

Basado en las respuestas del usuario, crea una contraseña segura aleatoria que cumple con los criterios especificados.

Muestra la contraseña generada como resultado.

Este ejercicio le permite al usuario personalizar la fortaleza y complejidad de su contraseña según sus preferencias, lo que es esencial para mantener la seguridad en línea.

Calculadora de edad de perros: Pide al usuario que ingrese la edad de su perro en años y calcula su edad equivalente en años humanos. La fórmula comúnmente utilizada es diferente para los primeros dos años y luego para el resto.

Simulador de ruleta: Simula un juego de ruleta en el que el usuario apuesta un monto en un número o color específico. Genera un número aleatorio y determina si el usuario gana o pierde.

Comienza por preguntar al usuario cuánto dinero quiere apostar en el juego de ruleta.

Luego, pregunta al usuario si quiere apostar a un número específico o a un color (rojo o negro).

Si el usuario elige apostar a un número, solicita que ingrese el número en el que quiere apostar (de 1 a 36, ya que son los números en una ruleta estándar).

Si el usuario elige apostar a un color, solicita que indique si desea apostar al color rojo o negro.

Genera un número aleatorio entre 1 y 36 para simular el resultado de la ruleta.

Comprueba si el resultado coincide con la apuesta del usuario (ya sea número o color).

Si coincide, el usuario gana un premio equivalente a su apuesta.

Si no coincide, el usuario pierde la apuesta.

Muestra el resultado del juego, indicando si el usuario ganó o perdió y cuánto dinero tiene después del juego.
