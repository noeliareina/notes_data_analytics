```python
# Añadir un año a la experiencia de cada trabajadora
# iteramos
for diccionario in personal["workers"]:
    print(diccionario)
```

```python
for diccionario in personal["workers"]:
    print(diccionario["experiencia"])
```

```python
for diccionario in personal["workers"]:
    diccionario["experiencia"] += 1
    diccionario["edad"] += 1
    print(diccionario)
```

```python
# No
# Es importantísimo fijarnos en estas lógicas para encontrar nuestros caminos
```

- 1 año = junior 
- 2 o más = mid 
- más de 4 senior

```python
for d in personal["workers"]:
    if d ["experiencia"] == 1:
        d ["nivel"] = "Junior" 

    elif d["experiencia"] >= 4:
        d ["nivel"] = "Senior" 

    else:
        d["nivel"] = "Mid"
    
```
