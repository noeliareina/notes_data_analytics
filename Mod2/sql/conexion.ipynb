```python
cnx = mysql.connector.connect(  
    host='localhost',
    user='root',
    password='AlumnaAdalab',    
    auth_plugin='mysql_native_password'
)

#iniciamos el cursor
mycursor = cnx.cursor()
```

```python
query = """CREATE SCHEMA tienda_prueba2;"""
```

```python
# creo la conexión que antes he cerrado, se ha vuelto a crear en workbench el esquema o base de datos
mycursor.execute(query)
```

```python
query_creacion = """CREATE TABLE cliente(
	                    id INT AUTO_INCREMENT,
                        nombre VARCHAR(10),
                        PRIMARY KEY (id));"""
```

```python
mycursor.execute(query_creacion)
```

```python
query_insercion = """INSERT INTO cliente (nombre)
VALUES ('Ana');"""
```

```python
mycursor.execute(query_insercion)
```

```python
cnx.commit() #confirmamos los cambios en la base de datos porque hemos hecho una insercion, modificacion de los datos
```

```python
query_insercion = """INSERT INTO cliente (nombre)
    VALUES (%s)""" # significa que ahínco vamos a poner un valor, como utilizar f-strings
```

```python
cliente = ["Carlos"] #importante que sea una lista

mycursor.execute(query_insercion, cliente)

cnx.commit()
```

```python
query_seleccion = """INSERT INTO cliente (id, nombre)
    VALUES (%s, %s)"""
```

```python
clientes = [(3, "Olivia"), (4, "Javier")]
```

```python
# EXECUTE MANY PARA INSERTAR VARIOS REGISTROS A LA VEZ
mycursor.executemany(query_seleccion, clientes) # paso la query de insercción y la lista de tuplas con los valores a insertar
# confirmamos los cambios
cnx.commit()
```

HAY VECES QUE EN VEZ DE UTILIZAR EL EXECUTEMANY UTILIZAMOS EL EXECUTE PERO LO HACEMOS EN BUCLE UTILIZANDO TRY...EXCEPT PARA CAPTURAR ERRORES. ASÍ PODEMOS SABER QUÉ REGISTROS HAN DADO ERROR Y CUALES NO.
Así personalizaremos los errores, lo que se llama elevar el error (raise) en programaicón. Por ejemplo, si tenemos un CSV con 1000 registros y 10 de ellos dan error, podemos capturar esos errores y guardarlos en una lista para luego revisarlos y corregirlos.
