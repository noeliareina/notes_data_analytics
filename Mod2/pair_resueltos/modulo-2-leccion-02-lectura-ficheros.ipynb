1. Hacer las importaciones necesarias para abrir un archivo de tipo "csv" en Pandas.

```python
import pandas as pd
import json
```

2. Abrir el archivo, creando un Dataframe. Aseguraros de que no incluye la columna "Unnamed: 0".

```python
df = pd.read_csv("videogames.csv", index_col=0)
df.head()
```

3. Guardar el Dataframe en formato Excel y después crear un Dataframe con el Excel. ¿EL Dataframe es similar al que teníais antes?

```python
df.to_excel('videogames.xlsx', index=False)
df_excel = pd.read_excel('videogames.xlsx')
df_excel.head(5)
```

4. Crear una carpeta llamada Files y guardar el mismo archivo en esa carpeta en formato csv .

```python
df_excel.to_csv('Files/videogames.csv', index=False)
```

6. Abre el json y haz las modificaciones necesarias para poder abrirlo correctamente en Pandas.

```python
# guardamos la ruta de nuestro archivo en una variable
json_path = 'Files/personal.json'

# Leer el JSON en un diccionario
with open(json_path, 'r') as f:
    data = json.load(f)

# 'workers' es la clave principal del JSON, entonces necesitamos acceder a esa lista
workers_data = data['workers']

# Normalizar los datos
df_workers = pd.json_normalize(workers_data)

# Mostrar las primeras filas del DataFrame para verificar
df_workers.head()
```

7. Guardar el json modifiado y abrirlo en Pandas para que tenga formato de tabla.

```python
df_workers.to_csv('Files/personal.csv', index=False)

df_personal = pd.read_csv("Files/personal.csv")

df_personal.head()
```
