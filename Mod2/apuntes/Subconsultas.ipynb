## 12. SQL Subconsultas
### 1. Subconsultas
Una consulta que se ejecuta dentro de otra consulta. Imagina que estás haciendo una gran pregunta (consulta principal) y dentro de esa pregunta, haces otra más pequeña (subconsulta) para obtener una respuesta que te ayude.
Se usan cuando necesitamos datos adicionales de otra tabla o resultado para completar nuestra consulta principal.
EJEMPLO
Si quisiéramos obtener los nombres de las empleadas que están trabajando en el proyecto 2, podemos hacer lo siguiente:

Primero, encontramos los id_empleada que están en ese proyecto.

Luego, usamos esa información para mostrar los nombres de esas empleadas.

Esto lo haríamos con una subconsulta.

Las subconsultas se suelen usar con palabras clave como IN, NOT IN, ANY, ALL y EXISTS.

- IN: Para seleccionar valores que coinciden con lo que devuelve la subconsulta (por ejemplo, empleadas que están en un proyecto).

- NOT IN: Para seleccionar los que no coinciden (empleadas que no están en algún proyecto).

- ANY y ALL: Para hacer comparaciones (por ejemplo, si una empleada tiene un salario mayor que el de cualquiera o todas las empleadas en otro país).

- EXISTS: Para verificar si algo existe en los resultados de una subconsulta (por ejemplo, si hay registros de un proyecto activo).

### 2. Subconsultas Correlacionadas
La subconsulta depende de cada fila de la consulta principal. Es decir, la subconsulta se ejecuta una vez por cada fila de la consulta principal.

Es como si la consulta principal y la subconsulta estuvieran teniendo una conversación constante: la consulta principal le pregunta algo a la subconsulta y luego decide qué hacer con los resultados.

Ejemplo de subconsulta correlacionada:

Imaginemos que queremos encontrar a las empleadas que ganan más que el salario promedio de las demás en su mismo país.
Para hacer esto:

- La consulta principal recorre las filas de la tabla de empleadas.

- Para cada empleada, la subconsulta calcula el salario promedio de las empleadas de su país.

Luego, la consulta principal compara el salario de esa empleada con el promedio calculado.

La clave aquí es que la subconsulta está "correlacionada" porque depende de los datos de cada empleada que se está procesando en la consulta principal.

¿Cuándo usarías una subconsulta correlacionada?
Este tipo de subconsultas es útil cuando necesitas comparar datos dentro de un grupo específico o contexto, como:

Comparar salarios dentro de un país, departamento o equipo.

Buscar valores relacionados con una categoría específica en tiempo real.

Hacer cálculos que cambian dependiendo del registro que estás evaluando en la consulta principal.
