**1 - Sacamos información de los jugadores**

El nombre, apellido y equipo de cada uno de los jugadores que tenemos en la NBA. Para eso tendrás que usar el siguiente endpoint `https://api.balldontlie.io/v1/players`. Extrae información de 200 jugadores. Devuelve los resultados en un DataFrame.

NOTA: recuerda que la API solo nos da 25 jugadores por defecto y un maximo de 100 por peticion!

```python
# ahora definamos lo que vamos a necesitar, la api_key, las paginas, la cantidad de jugadores, la cantidad de paginas, etc

api_key = f"6c61a37d-eb6a-4dac-aa3f-1c14feac6e4b"  
per_page = 100  # Número de jugadores por página (máximo permitido es 100 segun la API)
total_jugadores = 200  # Total de jugadores que queremos traer en total
pages = total_jugadores // per_page  # Calcular el número de páginas necesarias
```

Para que necesitamos las paginas? fijense en los parametros dados por la API, donde nosotras podemos definir la cantidad de jugadores por pagina que podemos tomar (25 por defecto y un maximo de 100).
Ahora bien, si consultamos la documentacion, en el apartado "pagination" habla sobre las paginas, esto es un parametro que me permite a mi elegir o ir saltando de pagina en pagina haciendo las consultas, entonces si podemos obtener por defecto 25 jugadores, y un maximo de 100 por consulta, podriamos intentar tener 100 jugadores en la primera pagina y los otros 100 en la segunda para completar el ejercicio no? eso haremos!

```python
# Inicializar el diccionario para almacenar la información de los jugadores
jugadores = {"nombres": [], "apellidos": [], "equipo": []}


for page in range(1, pages + 1): # este ciclo va consultando por cada pagina....ya definido anteriormente
    res_jugadores = requests.get(f"http://api.balldontlie.io/v1/players", 
                                 headers={'Authorization': api_key}, 
                                 params={'next_cursor': page, 'per_page': per_page}) # losparametros nos ayudan a darle forma a la consulta
    
    print(f"La llamada a la API para la página {page} nos ha dado una respuesta de tipo: {res_jugadores.status_code}")
    
    if res_jugadores.status_code != 200:  # Si la respuesta no fue exitosa
        print(f"El motivo por el que la llamada falló es {res_jugadores.reason}")
        break
    else:
        res_jugadores = res_jugadores.json()  # Convertir la respuesta a JSON

        # Llenar el diccionario con la información de los jugadores
        for jugador in res_jugadores["data"]:
            jugadores["nombres"].append(jugador["first_name"])
            jugadores["apellidos"].append(jugador["last_name"])
            jugadores["equipo"].append(jugador["team"]["full_name"])    

# Crear un DataFrame a partir del diccionario "jugadores"
df_jugadores = pd.DataFrame(jugadores)

# Mostrar el DataFrame
df_jugadores
```

**2 - Sacamos información de los equipos**

Extrae información de todos los equipos, con la informacion que estos mismos traen,puedes encontrar la conferencia a la que pertenece, la ciudad, el nombre y su abreviacion. En este caso ustedes deben buscar por dentro de la pagina cual es la API que corresponderia a esta busqueda!!

```python
equipos = {"conference": [], "name": [], "city": [], "abbreviation": []} # Armando mi diccionario donde guardare la informacion

res_equipos = requests.get(f"http://api.balldontlie.io/v1/teams", headers = {'Authorization': api_key}) # cada una pone la api_key que haya generado

# Verificar si la solicitud fue exitosa
if res_equipos.status_code == 200:
    # Convertir la respuesta JSON en un diccionario de Python
    res_equipos = res_equipos.json()
    
    # Iterar sobre los equipos y agregar información al diccionario de equipos
    for equipo in res_equipos["data"]:
        equipos["conference"].append(equipo["conference"])
        equipos["name"].append(equipo["full_name"])
        equipos["city"].append(equipo["city"])
        equipos["abbreviation"].append(equipo["abbreviation"])
    
    df_equipos = pd.DataFrame(equipos) # creando un dataframe a partir de mi diccionario "equipos"
    
    # En lugar de imprimir el DataFrame, simplemente guárdalo en una variable
else:
    print("Error al obtener los datos de los equipos.")

df_equipos # mostrando los resultados
```
