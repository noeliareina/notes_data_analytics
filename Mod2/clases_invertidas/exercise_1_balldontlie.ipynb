```python
import requests
import pandas as pd

# URL base de la API
url_players = "https://api.balldontlie.io/v1/players"

# Definimos cuántos jugadores queremos en total y por página
total_jugadores = 200
per_page = 100  # Máximo permitido por la API
pages = total_jugadores // per_page  # Calculamos cuántas páginas necesitamos

# Almacenaremos los jugadores obtenidos
data_players = []

# Iteramos sobre las páginas necesarias para obtener los jugadores
for page_num in range(1, pages + 1):
    print(f"Llamando a la API para la página {page_num}...")

    # Realizamos la petición a la API sin headers, con paginación correcta
    response = requests.get(url_players, headers={'Authorization': api_key}, params={'page': page_num, 'per_page': per_page})

    # Validamos si la respuesta es correcta (código 200)
    if response.status_code != 200:  
        print(f"Error al obtener datos. Código de estado: {response.status_code}")
        print(f"Motivo: {response.reason}")
        break
    
    try:
        # Intentamos convertir la respuesta a JSON
        data = response.json()
        
        # Recorremos cada jugador en la respuesta JSON
        for player in data['data']:
            data_players.append({
                'Nombre': player['first_name'],
                'Apellido': player['last_name'],
                'Equipo': player['team']['full_name']
            })

    except ValueError as e:
        print(f"Error al decodificar la respuesta JSON: {e}")
        break

# Convertimos los datos a un DataFrame de Pandas
df_jugadores = pd.DataFrame(data_players)

# Mostramos las primeras filas del DataFrame
print(df_jugadores.head())

# Verificamos la cantidad de jugadores obtenidos
print(f"Total de jugadores obtenidos: {len(df_jugadores)}")
```
